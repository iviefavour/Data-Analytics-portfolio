# **● What are the most popular product categories based on sales?**

```python
Popular_Category = alpha.groupby(['Category'])["Sales"].sum()
Popular_Category = Popular_Category.reset_index(name = "Sales").sort_values(by = "Sales", ascending= False)

print(Popular_Category)
Popular_Category.plot(kind = "bar")
```

# **● Can you identify the top 5 selling products?**

```python
top_5_products = alpha.groupby(["Product Name","Category"])["Sales"].sum()
top_5_products = top_5_products.reset_index(name = "Sales").sort_values(by = ["Sales"],ascending= False)
print(top_5_products[:5])
top_5_products[:5].plot(kind = "bar")
```

# **● Calculate the total profit for each order.**

```python
alpha["Cost"] = alpha.Sales - 6
alpha["Profit"] = alpha.Sales - alpha.Cost
Total_prof = pd.pivot_table(data = alpha, columns = "Order ID", values = "Profit", aggfunc = "sum").T
Total_prof.sort_values(by = "Profit", ascending = False)[:7]
```

# **What is the average profit margin %**

```python
alpha["ProfitMargin(%)"] = (alpha.Profit/alpha.Sales)*100
print("Average Profit Margin % =","$",alpha["ProfitMargin(%)"].mean())
```

# **● Identify the top 5 orders with the highest profit**

```python
alpha["Cost"] = alpha.Sales - 6
alpha["Profit"] = alpha.Sales - alpha.Cost
Total_prof = pd.pivot_table(data = alpha, columns = "Order ID", values = "Profit", aggfunc = "sum").T
Total_prof.sort_values(by = "Profit", ascending = False)[:5]
```
