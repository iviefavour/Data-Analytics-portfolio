```python
(5) Difference between Python list and NumPy array:

#Python List
General-purpose container → can hold anything: numbers, strings, even a mix of types ([1, "hello", 3.5] is totally fine).
Dynamic → size can change easily (append, remove, etc.).
Operations → element-wise math doesn’t work out-of-the-box. Example:
    
    [1, 2, 3] * 2   # → [1, 2, 3, 1, 2, 3] (repeats the list, not multiplies values)

Speed → slower for heavy numerical computations because it’s a higher-level structure.

#NumPy Array
Specialized container → designed for numerical data (all elements usually of the same type for efficiency).
Fixed type & size → takes less memory and allows for faster processing.
Operations → supports vectorized math. Example:
    
    import numpy as np
arr = np.array([1, 2, 3])
arr * 2   # → array([2, 4, 6]) (does actual math, not repetition)

Speed → blazing fast for numerical computations because it’s implemented in C under the hood.
```

```python
# (10) Generate random array of integers between 10 and 50:

import numpy as np
# For a single integer
random_int = np.random.randint(10, 51)
print("random_int:",random_int) 

# For an array of random integers
random_array = np.random.randint(10, 51, size=(3, 4))
print("random_array:",random_array)
```
