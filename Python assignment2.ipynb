(b) An employee resumed his office work at 8:15am and left the office at 17:23pm. Write a program to show
  How long he spent in the office

```python
from datetime import datetime

start = datetime.strptime("8:15 AM","%I:%M %p")
end = datetime.strptime("5:23 PM","%I:%M %p")

diff = end-start
print("time spent in the office:",diff)
```
