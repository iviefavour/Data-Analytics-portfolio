9. What is the difference between copy and view? with example

The difference between copy and view is that the changes made in copy does not affect the original while changes made in view affects the original

```python
# example of copy

import numpy as np
arr = np.array([2,4,6,8,10])
x = arr.copy()
arr[0]=20

print(arr)
print(x)
```

```python
# example of view

import numpy as np
arr = np.array([2,4,6,8,10])
x = arr.view()
arr[0]=20

print(arr)
print(x)
```
